```python
import segyio
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from shutil import copyfile
from skimage import exposure
```

```python
filename = 'Decatur_IBDP_GM1_3DVSP_Final_Migration_Depth_09-05-19.sgy'
VSP = 1-segyio.tools.cube(filename)
```

```python
print ('VSP type is: ' + str(VSP.__class__.__name__))
```

```python
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot()
sim = ax.imshow(VSP[50,:,:], cmap='gray_r');
fig.colorbar(sim, ax=ax)
ax.set_xticks([])
ax.set_yticks([])
ax.invert_xaxis()
```

```python
with segyio.open(filename, "r") as segyfile:
    # Print inline and crossline ranges
    print('VSP Inline range: ' + str(np.amin(segyfile.ilines)) + ' - ' +str(np.amax(segyfile.ilines))) 
    print('VSP Crossline range: ' + str(np.amin(segyfile.xlines)) + ' - ' +str(np.amax(segyfile.xlines)))
```
